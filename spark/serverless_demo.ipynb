```python
from pyspark.sql.functions import udf
```

```python
@udf(returnType="int")
def get_content_length(url: str):
    from library_for_spark.user_defined_function import page_content_length

    return page_content_length(url)
```

```python
df = spark.createDataFrame(
    [("https://www.google.com", 1), ("https://www.github.com", 2)], ["url", "id"]
)
df.show()
```

```python
df.withColumn("content_length", get_content_length("url")).show()
```
